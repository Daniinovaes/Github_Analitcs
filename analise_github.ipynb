Um dos indicadores mais valiosos de atividade e eficiência é a dinâmica das Pull Requests (PRs). PRs são essencialmente propostas de mudanças em um código, que são então revisadas, possivelmente ajustadas, e finalmente incorporadas (ou "mergeadas") ao código principal. Analisar as PRs abertas e fechadas pode nos fornecer insights significativos sobre a eficiência e a produtividade de uma equipe de desenvolvimento.

# O Ciclo de Tempo de um pull request 

Tempo entre o primeiro commit e a fusão do pull request. 
É a soma dos seguintes componentes:

<ol>
    <li>Tempo de Desenvolvimento: O tempo que o autor levou para entender os requisitos e produzir a primeira implementação funcional. Um longo tempo de desenvolvimento pode indicar problemas de planejamento ou de expertise.</li>
    <li>Tempo de Resposta: O tempo que um revisor levou para começar a analisar a alteração de código. Um longo tempo de resposta pode destacar problemas de trabalho em equipe ou silos de conhecimento.</li>
    <li>Tempo de Revisão: O tempo que o pull request levou para ser aprovado. Isso pode envolver várias idas e vindas entre o autor e o revisor. Um longo tempo de revisão pode revelar deficiências de processo e expertise.</li>
    <li>Tempo de Integração: O tempo que a alteração de código levou para ser integrada após ter sido aprovada. Um longo tempo de integração pode revelar problemas da organização com priorização e dependência externa.</li>
</ol>

## Por que o Ciclo de Tempo é importante?
o Ciclo de Tempo é usado para detectar ineficiências de processo, para que os gerentes possam mitigá-los antes que se tornem sistêmicos. Como regra geral, boas equipes de engenharia frequentemente têm um Ciclo de Tempo inferior a 48 horas, uma vez que qualquer problema mencionado acima arrastaria o Ciclo de Tempo além de 48 horas. Isso torna o Ciclo de Tempo a métrica geral para potenciais problemas que precisam ser abordados.

As informações que podemos extair do quadro de Time Cycle são:

<ol>
    <li>Melhoria na eficiência da revisão de código: A porcentagem de PRs submetidas à revisão aumentou de 17,8% em 2021 para 40,19% em 2023. Esse aumento na frequência de revisões é um indicativo claro de um processo de garantia de qualidade mais robusto e eficiente.</li>
    <li>Melhoria na eficiência geral do processo de desenvolvimento: Ao analisarmos o tempo de ciclo médio, vemos um decréscimo progressivo ao longo dos anos. Esse tempo, que engloba todas as etapas do processo, desde o commit inicial até o merge, caiu de 22 dias em 2021 para 7 dias em 2023.</li>

    <li>Processo de finalização do desenvolvimento mais eficiente: O tempo médio de fechamento para merge tem diminuído de forma constante.</li>

O tempo médio da revisão ao fechamento apresentou uma variação ao longo dos anos, aumentando de 2021 para 2022, mas diminuindo drasticamente em 2023. ## Análise de mudanças no codigo

Work Breakdown para mudanças no código, uma técnica valiosa, nos permite entender a natureza das mudanças, identificar padrões de trabalho, avaliar a produtividade da equipe e fornecer estimativas de tempo precisas para tarefas específicas.

Nessa análise, dividimos as mudanças de código em três categorias: New Code (novas funcionalidades), Refactoring (melhorias no código existente sem alterar a funcionalidade) e Rework (correção de erros). Isso fornece uma visão clara da distribuição de esforços, permitindo uma gestão mais eficaz do tempo e dos recursos.

Para aplicar essa técnica aos repositórios da OpenAI existiu um desafio: a falta de um padrão convencional nas descrições dos commits. foi adotado a seguinte regra: commits apenas com adições de código são considerados New Code, commits com a palavra "refactoring" ou seus derivados em sua descrição são categorizados como Refactoring, e o restante das mudanças como Rework.

Caso houvesse a adoção de um padrão como o "Conventional Commits" poderia tornar essa segregação mais precisa, estabelecendo uma linguagem comum e facilitando a análise de dados. Isso também melhora a comunicação, fornece uma melhor compreensão das alterações de código e permite previsões mais precisas para futuras mudanças. Como consequência, esse padrão iria aprimorar a confiabilidade e eficiência do processo de Work Breakdown.

No geral, o rework domina o cenário, representando uma média de 64% do trabalho total. A adição de novos códigos vem em segundo lugar, com 32,6%, e a refatoração compõe uma porção menor, com apenas 2,51%.

Entretanto, quando olhamos para a evolução desse work breakdown nos últimos três anos, percebemos algumas oscilações significativas. Em 2021, o rework representou uma parcela esmagadora de 81,7% do trabalho, deixando apenas 17,3% para novos códigos e uma quantidade mínima de 1,01% para refatoração. No ano seguinte, 2022, houve uma inversão notável, com novos códigos ultrapassando o rework por uma margem pequena (53,1% contra 45,5%). A refatoração manteve-se relativamente constante, com um ligeiro aumento para 1,4%. Já em 2023, a tendência voltou a favorecer o rework, que subiu para 76,3%, enquanto os novos códigos diminuíram para 22,9%. A refatoração, mais uma vez, manteve-se em um nível muito baixo, com apenas 0,8%.

Essas tendências nos levam a questionar o equilibrio dos esforços de desenvolvimento. Embora o rework seja uma parte necessária do desenvolvimento de software, sua predominância sugere que podemos estar gastando muitos recursos corrigindo erros ou refazendo o trabalho. Por outro lado, a proporção relativamente baixa de refatoração indica que podemos não estar dedicando tempo suficiente para melhorar a qualidade do nosso código existente. Isso é essencial para manter a manutenibilidade do código e minimizar a necessidade de rework no futuro.

Essas observações destacam a importância de uma abordagem equilibrada para o desenvolvimento de software. Precisamos não só adicionar novas funcionalidades e corrigir erros, mas também dedicar tempo para melhorar a qualidade do nosso código existente. Ao fazer isso, podemos melhorar a eficiência do nosso processo de desenvolvimento e garantir a qualidade dos nossos produtos.
